```python
#splitting the data into categorical and numerical columns for analysis
categorical_cols = data.select_dtypes(include=['object']).columns
numeric_cols = data.select_dtypes(include=[np.number]).columns
```

```python
for col in categorical_cols:
    sns.countplot(x=col, data=data, hue='Y', order=data[col].value_counts().index) 
    plt.title(f"Count plot for {col}")
    plt.xticks(rotation=90)
    plt.show()

# Direction "No Urgent Place" has the highest number of responses for the target variable 1 close to 4k 
# Sunny has the highest number of responses and target variable 1 close to 6k
# with friends people are more likely to use the coupon
# if coupon were to expire in 1 day (rather than few hours) people are more likely to use the coupon 
```

```python
for col in categorical_cols:
    ctab = pd.crosstab(data[col], data['Y'], normalize='index')*100
    print(f"\n=========={col}==========\n")
    print(ctab)

    #coupon type 
    #73.5% used it for take away 
    # 70% used it for a restaurant less than $20 (inexpensive)

    #age
    # younger population used the coupon more than the older population

    #marital status 
    # single people used the coupon more 
    
```
